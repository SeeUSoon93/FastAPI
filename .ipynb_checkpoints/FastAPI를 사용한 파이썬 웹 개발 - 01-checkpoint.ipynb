### 1.2 virtualenv를 사용한 개발 환경 구축
- 가상 환경에서 애플리케이션을 개발하면 특정 패키지를 시스테메 전역으로 설치하지 않아도 됨
- 서로 다른 애플리케이션을 동시에 개발할 때 발생할 수 있는 충돌을 피할 수 있음

#### 가상환경 생성
- venv 모듈을 사용해 가상 환경 생성 (mkdir todos && cd todos)
- todos라는 폴더를 만들고 이 폴더 안에 가상 환경 생성하기 (python3 -m venv venv)
- 가상 환경 활성화 (source venv/bin/activate)
- 가상 환경 비활성화 (deactivate)

### 1.3 pip을 사용한 패키지 관리

#### pip 설치
- 파이선 설치 시 함께 설치 됨. 다음 명령을 사용해 pip이 설치됐는지 확인 (python3 -m pip list)
#### 기본명령
- fastapi설치 (pip install fastapi)
- 모든 패키지 목록을 파일로 저장 (pip freeze > requirements.txt)
- requirements에 있는 패키지를 모두 설치. 주로 배포할 때 사용 (pip install -r requirements.txt)

### 1.4 도커 설정
- 애플리케이션은 여러 계층으로 구성(데이터베이스, 웹 서버 등) > 애플리케이션을 배포하려면 하나로 묶어야 한다 > 도커를 사용해 컨테이너화
- 도커파일과 도커 구성파일을 사용하면 애플리케이션 이미지를 업로드하거나 공유할 필요없음.(도커 허브에 저장했다가 다시 추출)

#### 도커파일
FROM python:3.8  
WORKDIR /usr/src/app           ① 작업 디렉터리 설정  
ADD . /usr/src/app             ② 로컬 디렉터리 파일을 작업 디렉터리로 복사  
CMD ["python","hello.py"]      ③ 명령 실행  

### 1.5 간단한 FastAPI 애플리케이션 개발
- 가상 환경 활성화 후 라이브러리와 패키지 설치
- fastapi : 애플리케이션 구축을 위한 프레임워크
- uvicorn : 애플리케이션 실행을 위한 비동기 방식 서버 게이트웨이 인터페이스

- api.py라는 파일을 생성해 인스턴스 생성( vi api.py)
- i를 입력하여 INSERT상태로 바꾸기. esc를 누르고 :wq를 입력해 저장하고 편집 종료

from fastapi import FastAPI  
  
app = FastAPI()  
  
@app.get("/")  
async def welcome() -> dict:  
    return {  
            "message": "Hello World"  
            }  

- uvicorn을 사용해 애플리케이션 실행 (uvicorn api:app --port 8000 --reload)
- 실행시 지정하는 인수
    - file:instance : FastAPI 인스턴스가 존재하는 파이썬 파일과 FastAPI 인스턴스를 가지고 있는 변수 지정
    - --port PORT : 애플리케이션에 접속할 수 있는 포트 번호
    - --reload : 선택적 인수, 파일이 변경될 때마다 애플리케이션 재시작

- 새 터미널을 열어 GET 요청 보내기 (curl http://localhost:8000/)
